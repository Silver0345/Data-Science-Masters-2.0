### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px
```

```python
titanic = sns.load_dataset('titanic')
titanic.columns
```

```python
fig = px.scatter(titanic, x='age', y='fare')
fig.show()
```

### Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
tips = px.data.tips()
fig = px.box(tips, x='tip')
fig.show()
```

### Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
fig = px.histogram(tips, x='sex', y='total_bill', pattern_shape='smoker', color='day')
fig.show()
```

### Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
import plotly.figure_factory as ff
```

```python
iris = px.data.iris()
```

```python
fig = ff.create_scatterplotmatrix(iris, diag='histogram',
                        index='species', colormap="Viridis",
                        colormap_type='cat', height=800, width=800)

fig.update_layout(title="Scatter Matrix Plot of Iris Dataset")

fig.show()
```

### Q5. What is Distplot? Using Plotly express, plot a distplot

**A Distplot:** is short for distribution plot, it is a data visualization tool used to display the distribution of a univariate (single-variable) dataset.

```python
fig = px.histogram(tips, x='total_bill', nbins=10, marginal='box')
fig.show()
```
